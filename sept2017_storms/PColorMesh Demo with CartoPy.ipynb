```python
import matplotlib
from matplotlib import pyplot as plt
import cartopy.crs as ccrs

import numpy as np

%matplotlib inline
```

```python
lats = np.arange(-90,90,1)
lons = np.arange(-180,180,1)
```

```python
LATS,LONS=np.meshgrid(lats,lons)
```

```python
data = np.arange(lats.size*lons.size)
```

```python
data.shape = (lons.size,lats.size)
```

```python
data
```

```python
data.shape
```

```python
fig = plt.figure(figsize=(15,8))
ax  = fig.add_subplot(111,projection=ccrs.PlateCarree())

ax.coastlines()
ax.gridlines(draw_labels=True)

ax.pcolorfast(lons,lats,data.T)
#ax.pcolormesh
#ax.contourf

ax.set_xlim(-130,-50) # Longitude
ax.set_ylim(25,65) # Latitude

plt.draw()
```
